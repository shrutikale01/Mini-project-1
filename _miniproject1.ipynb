```python
#3. **User Engagement Analysis**
ratings = open(r"C:\Users\Dell\OneDrive\Desktop\Dataset\ratings.dat")
user_rating_count = dict()  

users = open(r"C:\Users\Dell\OneDrive\Desktop\Dataset\users.dat")
user_demographics = dict()  

for line in users:
    line = line.strip()
    columns = line.split('::')
    userId = int(columns[0])  
    gender = columns[1]  
    age = int(columns[2])  
    occupation = columns[3]  
    user_demographics[userId] = {'Gender': gender, 'Age': age, 'Occupation': occupation}

for line in ratings:
    line = line.strip()
    columns = list(map(int, line.split('::')))  
    userId = columns[0]
    rating = columns[2]

    if userId in user_rating_count:
        user_rating_count[userId] += 1
    else:
        user_rating_count[userId] = 1

sorted_user_activity = sorted(user_rating_count.items(), key=lambda x: x[1], reverse=True)

print("Most Active Users :")
for userId, count in sorted_user_activity[:10]:
    user_info = user_demographics.get(userId, {})
    print(f"UserID: {userId}, Ratings: {count}, Gender: {user_info.get('Gender')}, Age: {user_info.get('Age')}, Occupation: {user_info.get('Occupation')}")

gender_rating_sum = dict()
gender_rating_count = dict()

age_rating_sum = dict()
age_rating_count = dict()

occupation_rating_sum = dict()
occupation_rating_count = dict()

ratings.seek(0)  

for line in ratings:
    line = line.strip()
    columns = list(map(int, line.split('::')))  
    userId = columns[0]
    rating = columns[2]
    
    gender = user_demographics.get(userId, {}).get('Gender')
    if gender:
        if gender in gender_rating_sum:
            gender_rating_sum[gender] += rating
            gender_rating_count[gender] += 1
        else:
            gender_rating_sum[gender] = rating
            gender_rating_count[gender] = 1
    
    age = user_demographics.get(userId, {}).get('Age')
    if age:
        if age in age_rating_sum:
            age_rating_sum[age] += rating
            age_rating_count[age] += 1
        else:
            age_rating_sum[age] = rating
            age_rating_count[age] = 1
    
    occupation = user_demographics.get(userId, {}).get('Occupation')
    if occupation:
        if occupation in occupation_rating_sum:
            occupation_rating_sum[occupation] += rating
            occupation_rating_count[occupation] += 1
        else:
            occupation_rating_sum[occupation] = rating
            occupation_rating_count[occupation] = 1

print("\nAverage Rating by Gender:")
for gender in gender_rating_sum:
    average_rating = gender_rating_sum[gender] / gender_rating_count[gender]
    print(f"{gender}: {average_rating:.2f}")

print("\nAverage Rating by Age Group:")
for age in age_rating_sum:
    average_rating = age_rating_sum[age] / age_rating_count[age]
    print(f"Age {age}: {average_rating:.2f}")

print("\nAverage Rating by Occupation:")
for occupation in occupation_rating_sum:
    average_rating = occupation_rating_sum[occupation] / occupation_rating_count[occupation]
    print(f"Occupation {occupation}: {average_rating:.2f}")
```

```python
# 5. **Top Performers**
ratings = open(r"C:\Users\Dell\OneDrive\Desktop\Dataset\ratings.dat")
movies = open(r"C:\Users\Dell\OneDrive\Desktop\Dataset\movies.dat")

movie_ratings = {}  
movie_counts = {}   
movieId_genre = {}  
movieId_year = {}   

for line in movies:
    movieId, title, genres = line.strip().split('::')
    genres_list = genres.split('|')
    movieId_genre[int(movieId)] = genres_list
    year = title.split('(')[-1].split(')')[0] if '(' in title else 'Unknown'
    movieId_year[int(movieId)] = year

 
for line in ratings:
    userId, movieId, rating, _ = map(int, line.strip().split('::'))
    if movieId not in movie_ratings:
        movie_ratings[movieId] = 0
        movie_counts[movieId] = 0
    movie_ratings[movieId] += rating
    movie_counts[movieId] += 1

MIN_RATINGS = 50

movie_avg_ratings = {}
for movieId in movie_ratings:
    if movie_counts[movieId] >= MIN_RATINGS:
        movie_avg_ratings[movieId] = movie_ratings[movieId] / movie_counts[movieId]

top_movies = sorted(movie_avg_ratings.items(), key=lambda x: x[1], reverse=True)[:10]

highest_avg_rated_movies = []   
top_rated_movies = []           

print("Highest average rated movies and their genre\n")
for movieId, avg_rating in top_movies:
    title = movieId_year.get(movieId, 'Unknown Year')
    genres = ", ".join(movieId_genre.get(movieId, []))
    highest_avg_rated_movies.append((title, genres))
    print(f"Movie Name: {title}  Genre: {genres}")

print("\n*******************************************************************************************************\n")
print("Top rated movies and their genre\n")
for movieId, avg_rating in top_movies:
    title = movieId_year.get(movieId, 'Unknown Year')
    genres = ", ".join(movieId_genre.get(movieId, []))
    top_rated_movies.append((title, genres))
    print(f"Movie Name: {title}  Genre: {genres}")
```

```python
# 6. **Exploring Long Tail**
ratings = open(r"C:\Users\Dell\OneDrive\Desktop\Dataset\ratings.dat")
movies = open(r"C:\Users\Dell\OneDrive\Desktop\Dataset\movies.dat")

movie_ratings = {}
movie_counts = {}
movieId_genre = {}
 
 
for line in movies:
    movieId, title, genres = line.strip().split('::')
    movieId_genre[int(movieId)] = genres.split('|')

for line in ratings:
    _, movieId, _, _ = map(int, line.strip().split('::'))
    movie_counts[movieId] = movie_counts.get(movieId, 0) + 1
 
MIN_RATINGS = 50
LONG_TAIL_THRESHOLD = 10

long_tail_movies = {movieId: count for movieId, count in movie_counts.items() if count < LONG_TAIL_THRESHOLD}
popular_movies = {movieId: count for movieId, count in movie_counts.items() if count >= MIN_RATINGS}

print(f"Long tail movies (fewer than {LONG_TAIL_THRESHOLD} ratings): {len(long_tail_movies)}")
print("\nExample Long Tail Movies:")
for movieId in list(long_tail_movies.keys())[:10]:
    genres = ", ".join(movieId_genre.get(movieId, []))
    print(f"Movie ID: {movieId}  Genre: {genres}  Rating Count: {long_tail_movies[movieId]}")
    
print("\n*******************************************************************************************************")

print("\nPopular Movies:")
for movieId in list(popular_movies.keys())[:10]:
    genres = ", ".join(movieId_genre.get(movieId, []))
    print(f"Movie ID: {movieId}  Genre: {genres}  Rating Count: {popular_movies[movieId]}")
```
